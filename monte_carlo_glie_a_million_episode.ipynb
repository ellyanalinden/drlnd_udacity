### Part 1: MC Prediction

In this section, you will write your own implementation of MC prediction (for estimating the action-value function).  

We will begin by investigating a policy where the player _almost_ always sticks if the sum of her cards exceeds 18.  In particular, she selects action `STICK` with 80% probability if the sum is greater than 18; and, if the sum is 18 or below, she selects action `HIT` with 80% probability.  The function `generate_episode_from_limit_stochastic` samples an episode using this policy. 

The function accepts as **input**:
- `bj_env`: This is an instance of OpenAI Gym's Blackjack environment.

It returns as **output**:
- `episode`: This is a list of (state, action, reward) tuples (of tuples) and corresponds to $(S_0, A_0, R_1, \ldots, S_{T-1}, A_{T-1}, R_{T})$, where $T$ is the final time step.  In particular, `episode[i]` returns $(S_i, A_i, R_{i+1})$, and `episode[i][0]`, `episode[i][1]`, and `episode[i][2]` return $S_i$, $A_i$, and $R_{i+1}$, respectively.

````python
def generate_episode_from_limit_stochastic(bj_env):
    episode = []
    state = bj_env.reset()
    while True:
        probs = [0.8, 0.2] if state[0] > 18 else [0.2, 0.8]
        action = np.random.choice(np.arange(2), p=probs)
        next_state, reward, done, info = bj_env.step(action)
        episode.append((state, action, reward))
        state = next_state
        if done:
            break
    return episode
````

### Part 2: MC Control

In this section, you will write your own implementation of constant-$\alpha$ MC control.  

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.
- `policy`: This is a dictionary where `policy[s]` returns the action that the agent chooses after observing state `s`.

(_Feel free to define additional functions to help you to organize your code._)

````python
# this Q fc takes environment, the most recent estimate of the Q table, 
# the value of epsilon and the total number of action available to agent as input
# As output, it returns an episode which is just a sequence state-action-reward tuple 
def generate_episode_from_Q(env, Q, epsilon, nA):
    ''' generates an episode from the following the epsilon-greedy policy'''
    episode =[]
    state = env.reset()
    while True:
        # along the way the agent will use the Epsilon greedy policy to select actions
        # we implemented that using random.choice method from numpy
        # it takes input the set of possible actions and 
        # we also have to feed it the probabilities corresponding to the Epsilon greedy policy 
        # specifically, when selecting an action from a state we need to get the action probabilities corresponding to 
        # epsilon greedy action selection from that state. 
        # This is implemented in get_probs fc which takes input epsilon and number of actions available to agent, 
        # the row in the Q-table corresponding to the state that the agent has to select the action from
        action = np.random.choice(np.arange(nA), p=get_probs(Q[state], epsilon, nA)) \
                                    if state in Q else env.action_space.sample()
        next_state, reward, done, info = env.step(action)
        episode.append((state,action,reward))
        state = next_state
        if done:
            break
    return episode

def get_probs(Q_s, epsilon, nA):
    ''' obtains the action probabilities corresponding to epsilon-greedy policy'''
    policy_s = np.ones(nA) * epsilon / nA
    best_a = np.argmax(Q_s)
    policy_s[best_a] = 1 - epsilon + (epsilon / nA)
    return policy_s

# after we got the return episode, we update Q fc. To see how it ties to MC control...go to MC Control fc
def update_Q(env, episode, Q, alpha, gamma):
    '''updates the action-value fc estimate using the most recent episodes'''
    states, actions, rewards = zip(*episode)
    # prepare for discounting
    discounts = np.array([gamma**i for i in range(len(rewards)+1)])
    for i, state in enumerate(states):
        old_Q = Q[state][actions[i]]
        Q[state][actions[i]] = old_Q + alpha *(sum(rewards[i:]*discounts[:-(i+1)]) - old_Q)
    return Q
````

````python
def mc_control(env, num_episodes, alpha, gamma=1.0, eps_start=1.0, eps_decay=.99999, eps_min=0.05):
    nA = env.action_space.n
    # Q is initialized to an empty dictionary of arrays
    # We need to parse in the total number of actions that are in the environment. We keep track with variable nA.
    Q = defaultdict(lambda: np.zeros(nA))
    # set the value of initial epsilon. We start off the value of epsilon at some initialized value epsilon start.
    # eps_start = 1.0, then with episode we slightly decay the epsilon value by multipying it with eps_decay
    epsilon = eps_start
    # After Q initialization, we loop over episodes
    # With each episode we set the value of epsilon
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 1000 == 0:
            # keep track how many episodes completed out of the total
            print("\rEpisode {}/{}.".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODO: complete the function
        # set the value of epsilon
        # We don't want epsilon too small b'c we want to constantly ensure at least some small exploration throughout the process
        # Therefore, we don't want the epsilon value goes below eps_min which is 0.05
        # After epsilon value is updated, it is sent to generate_episode_from_Q fc 
        epsilon = max(epsilon*eps_decay, eps_min)
        # construct the epsilon greedy policy with respect the most recent estimate of the Q table
        # generate an episode by following epsilon_greedy policy
        episode = generate_episode_from_Q(env, Q, epsilon, nA)
        # update the action-value fc estimate using the episode
        # this is where the update-Q fc ties to the MC Control. 
        # After generating an episode, we parse the last estimate of the Q table into this update_Q fc which give us new values
        # of Q-table
        Q = update_Q(env, episode, Q, alpha, gamma)
    # determine the policy corresponding to the final action-value fc estimate
    # to get the corresponding optimal policy, for each state we just pick the action that maximizes the row corresponding to 
    # that state.
    policy = dict((k, np.argmax(v)) for k, v in Q.items())
    return policy, Q
````

As long as we run above algo long enough, it's guaranteed that this Q-table will correspond to optimal action-value fc

````python
# obtain the estimated optimal policy and action-value function
# we run till 500000 episodes
policy, Q = mc_control(env, 1000000, 0.03)
````
